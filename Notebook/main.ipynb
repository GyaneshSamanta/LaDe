```python
import os
import glob
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Set plotting style
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)

# =============================================================================
# 1. Define File Paths
# =============================================================================

# Change these paths if needed
delivery_path = r"C:\Users\mailg\OneDrive\Documents\GitHub\LaDe\LaDe\delivery"
pickup_path = r"C:\Users\mailg\OneDrive\Documents\GitHub\LaDe\LaDe\pickup"
road_path   = r"C:\Users\mailg\OneDrive\Documents\GitHub\LaDe\LaDe\road-network\roads.csv"

# Get list of delivery and pickup files (assuming naming pattern "delivery_*.csv" and "pickup_*.csv")
delivery_files = glob.glob(os.path.join(delivery_path, "delivery_*.csv"))
pickup_files = glob.glob(os.path.join(pickup_path, "pickup_*.csv"))

# For this example, we will load one sample from each category.
# (You can later use pd.concat() to combine all files in each category.)
sample_delivery_file = os.path.join(delivery_path, "delivery_cq.csv")
sample_pickup_file = os.path.join(pickup_path, "pickup_cq.csv")

# =============================================================================
# 2. Load the Data
# =============================================================================

# Load Delivery data (CSV is comma separated)
delivery_df = pd.read_csv(sample_delivery_file)
print("Delivery Data Sample:")
print(delivery_df.head())
print(delivery_df.info())

# Load Pickup data (CSV is comma separated)
pickup_df = pd.read_csv(sample_pickup_file)
print("\nPickup Data Sample:")
print(pickup_df.head())
print(pickup_df.info())

# Load Road network data (CSV is tab separated)
roads_df = pd.read_csv(road_path, sep="\t", encoding="utf-8")
print("\nRoad Network Data Sample:")
print(roads_df.head())
print(roads_df.info())

# =============================================================================
# 3. Exploratory Data Analysis and Visualizations
# =============================================================================

# ---------------------------
# 3.1. Road Network Analysis
# ---------------------------
# Visualize distribution of road classes (fclass)
plt.figure()
sns.countplot(data=roads_df, x="fclass", order=roads_df["fclass"].value_counts().index)
plt.title("Distribution of Road Classes")
plt.xlabel("Road Class")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Visualize distribution of maxspeed (if available)
if "maxspeed" in roads_df.columns:
    plt.figure()
    sns.histplot(data=roads_df, x="maxspeed", bins=20, kde=True, color='coral')
    plt.title("Distribution of Road Maxspeed")
    plt.xlabel("Maxspeed")
    plt.ylabel("Frequency")
    plt.show()

# ---------------------------
# 3.2. Delivery Data Analysis
# ---------------------------
# Check basic statistics and distribution
print("\nDelivery Data Description:")
print(delivery_df.describe())

# Assume delivery_df contains columns like 'delivery_time', 'lng', 'lat', 'courier_id'
# If delivery_time is not datetime, convert it
if delivery_df['delivery_time'].dtype == object:
    delivery_df['delivery_time'] = pd.to_datetime(delivery_df['delivery_time'], errors='coerce')

# Plot spatial distribution of deliveries
plt.figure()
sns.scatterplot(data=delivery_df, x="lng", y="lat", hue="courier_id",
                palette="tab20", legend=False, s=30, alpha=0.7)
plt.title("Spatial Distribution of Deliveries")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()

# Plot delivery time distribution (if column exists)
if "delivery_time" in delivery_df.columns:
    plt.figure()
    # For a time distribution, we can extract hour from delivery_time
    delivery_df["hour"] = delivery_df["delivery_time"].dt.hour
    sns.countplot(x="hour", data=delivery_df, palette="mako")
    plt.title("Hourly Distribution of Deliveries")
    plt.xlabel("Hour of Day")
    plt.ylabel("Number of Deliveries")
    plt.show()

# ---------------------------
# 3.3. Pickup Data Analysis
# ---------------------------
# Check basic statistics and distribution
print("\nPickup Data Description:")
print(pickup_df.describe())

# Convert pickup accept_time to datetime (assuming format like "08-14 07:57:00")
if pickup_df['accept_time'].dtype == object:
    pickup_df['accept_time'] = pd.to_datetime(pickup_df['accept_time'], format="%m-%d %H:%M:%S", errors='coerce')

# Plot spatial distribution of pickups
plt.figure()
sns.scatterplot(data=pickup_df, x="lng", y="lat", hue="courier_id",
                palette="tab20", legend=False, s=30, alpha=0.7)
plt.title("Spatial Distribution of Pickups")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()

# Plot pickup accept time distribution (extracting hour)
if "accept_time" in pickup_df.columns:
    pickup_df["hour"] = pickup_df["accept_time"].dt.hour
    plt.figure()
    sns.countplot(x="hour", data=pickup_df, palette="coolwarm")
    plt.title("Hourly Distribution of Pickup Accept Times")
    plt.xlabel("Hour of Day")
    plt.ylabel("Number of Pickups")
    plt.show()

# ---------------------------
# 3.4. Aggregated Analysis Across Multiple Files
# ---------------------------
# If you want to aggregate data from all 5 delivery files:
if delivery_files:
    all_delivery_df = pd.concat([pd.read_csv(f) for f in delivery_files], ignore_index=True)
    # Example: count deliveries by city if column "city" exists
    if "city" in all_delivery_df.columns:
        plt.figure()
        sns.countplot(data=all_delivery_df, x="city", order=all_delivery_df["city"].value_counts().index)
        plt.title("Delivery Count by City")
        plt.xlabel("City")
        plt.ylabel("Number of Deliveries")
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()

# Similarly, for pickup files:
if pickup_files:
    all_pickup_df = pd.concat([pd.read_csv(f) for f in pickup_files], ignore_index=True)
    if "city" in all_pickup_df.columns:
        plt.figure()
        sns.countplot(data=all_pickup_df, x="city", order=all_pickup_df["city"].value_counts().index)
        plt.title("Pickup Count by City")
        plt.xlabel("City")
        plt.ylabel("Number of Pickups")
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()

# ---------------------------
# 3.5. Courier-Level Aggregation and Clustering (for Delivery Data)
# ---------------------------
# Here we aggregate package-level delivery data to get a courier-level summary.
if set(["courier_id", "lat", "lng", "delivery_time"]).issubset(delivery_df.columns):
    courier_agg = delivery_df.groupby("courier_id").agg({
        "lat": "mean",
        "lng": "mean",
        "delivery_time": "count"  # count number of deliveries
    }).reset_index()
    courier_agg.rename(columns={"delivery_time": "num_deliveries"}, inplace=True)
    print("\nAggregated Courier-Level Data:")
    print(courier_agg.head())

    # Visualize spatial distribution of courier centroids (with size representing delivery count)
    plt.figure()
    sns.scatterplot(data=courier_agg, x="lng", y="lat", size="num_deliveries", legend=False, alpha=0.7, color='purple')
    plt.title("Courier Delivery Centroids")
    plt.xlabel("Average Longitude")
    plt.ylabel("Average Latitude")
    plt.show()

    # Clustering on courier-level data (using geographic centroids and delivery counts)
    features = courier_agg[["lat", "lng", "num_deliveries"]]
    scaler = StandardScaler()
    features_scaled = scaler.fit_transform(features)
    # Determine optimal clusters using the elbow method
    inertias = []
    K_range = range(1, 11)
    for k in K_range:
        km = KMeans(n_clusters=k, random_state=42)
        km.fit(features_scaled)
        inertias.append(km.inertia_)
    plt.figure()
    plt.plot(K_range, inertias, marker="o")
    plt.title("Elbow Method for Optimal K (Couriers)")
    plt.xlabel("Number of Clusters")
    plt.ylabel("Inertia")
    plt.xticks(K_range)
    plt.show()

    # For demonstration, choose k=3
    k_optimal = 3
    km = KMeans(n_clusters=k_optimal, random_state=42)
    courier_agg["cluster"] = km.fit_predict(features_scaled)
    plt.figure()
    sns.scatterplot(data=courier_agg, x="lng", y="lat", hue="cluster", palette="Set1", s=100, alpha=0.8)
    plt.title("K-Means Clustering of Couriers")
    plt.xlabel("Average Longitude")
    plt.ylabel("Average Latitude")
    plt.legend(title="Cluster")
    plt.show()

# =============================================================================
# End of Script
# =============================================================================
print("Visualizations generated successfully!")
```

```python
import os

def print_directory_structure(root_dir, indent=""):
    # List all items in the current directory
    try:
        items = os.listdir(root_dir)
    except PermissionError:
        print(indent + "[Permission Denied]")
        return

    for item in sorted(items):
        path = os.path.join(root_dir, item)
        if os.path.isdir(path):
            print(indent + "📁 " + item)
            # Recursively print the subdirectory structure
            print_directory_structure(path, indent + "    ")
        else:
            print(indent + "📄 " + item)

# Replace this with the path you want to scan
root_directory = r"C:\Users\mailg\OneDrive\Documents\GitHub\LaDe"
print_directory_structure(root_directory)
```

```python
import os
import glob
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Set global plotting style
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)

# =============================================================================
# 1. Define File Paths
# =============================================================================
delivery_path = r"C:\Users\mailg\OneDrive\Documents\GitHub\LaDe\LaDe\delivery"
pickup_path   = r"C:\Users\mailg\OneDrive\Documents\GitHub\LaDe\LaDe\pickup"
road_path     = r"C:\Users\mailg\OneDrive\Documents\GitHub\LaDe\LaDe\road-network\roads.csv"

delivery_files = glob.glob(os.path.join(delivery_path, "delivery_*.csv"))
pickup_files   = glob.glob(os.path.join(pickup_path, "pickup_*.csv"))

# For demonstration, use one sample file from each category.
sample_delivery_file = os.path.join(delivery_path, "delivery_cq.csv")
sample_pickup_file   = os.path.join(pickup_path, "pickup_cq.csv")

# =============================================================================
# 2. Load the Data
# =============================================================================

# Load Delivery data (comma separated)
delivery_df = pd.read_csv(sample_delivery_file)
# Load Pickup data (comma separated)
pickup_df = pd.read_csv(sample_pickup_file)
# Load Road network data (tab separated)
roads_df = pd.read_csv(road_path, sep="\t", encoding="utf-8")

# Convert time columns where needed
if delivery_df["delivery_time"].dtype == object:
    delivery_df["delivery_time"] = pd.to_datetime(delivery_df["delivery_time"], errors="coerce")

if pickup_df["accept_time"].dtype == object:
    # Format assumed: "08-14 07:57:00". You may need to adjust the format string if different.
    pickup_df["accept_time"] = pd.to_datetime(pickup_df["accept_time"], format="%m-%d %H:%M:%S", errors="coerce")

# =============================================================================
# 3. Descriptive Statistics & Correlation Matrices
# =============================================================================

# --- 3.1 Delivery Data Statistics ---
print("Delivery Data Descriptive Statistics:")
delivery_desc = delivery_df.describe(include="all")
print(delivery_desc)
# Save to CSV if needed: delivery_desc.to_csv("delivery_descriptive_stats.csv")

# Compute correlation matrix for numeric columns in delivery data
delivery_numeric = delivery_df.select_dtypes(include=[np.number])
if not delivery_numeric.empty:
    delivery_corr = delivery_numeric.corr()
    print("\nDelivery Data Correlation Matrix:")
    print(delivery_corr)
    
    plt.figure()
    sns.heatmap(delivery_corr, annot=True, cmap="coolwarm", fmt=".2f")
    plt.title("Delivery Data Correlation Matrix")
    plt.show()
    
    # Pairplot for selected numeric features (if enough numeric features exist)
    sns.pairplot(delivery_numeric.dropna(), diag_kind="kde", corner=True)
    plt.suptitle("Delivery Data Pairplot", y=1.02)
    plt.show()

# --- 3.2 Pickup Data Statistics ---
print("\nPickup Data Descriptive Statistics:")
pickup_desc = pickup_df.describe(include="all")
print(pickup_desc)
# Save to CSV if needed: pickup_desc.to_csv("pickup_descriptive_stats.csv")

pickup_numeric = pickup_df.select_dtypes(include=[np.number])
if not pickup_numeric.empty:
    pickup_corr = pickup_numeric.corr()
    print("\nPickup Data Correlation Matrix:")
    print(pickup_corr)
    
    plt.figure()
    sns.heatmap(pickup_corr, annot=True, cmap="coolwarm", fmt=".2f")
    plt.title("Pickup Data Correlation Matrix")
    plt.show()
    
    # Pairplot for numeric columns in pickup data
    sns.pairplot(pickup_numeric.dropna(), diag_kind="kde", corner=True)
    plt.suptitle("Pickup Data Pairplot", y=1.02)
    plt.show()

# --- 3.3 Road Network Data Statistics ---
print("\nRoad Network Data Descriptive Statistics:")
roads_desc = roads_df.describe(include="all")
print(roads_desc)
# Save to CSV if needed: roads_desc.to_csv("roads_descriptive_stats.csv")

roads_numeric = roads_df.select_dtypes(include=[np.number])
if not roads_numeric.empty:
    roads_corr = roads_numeric.corr()
    print("\nRoad Network Data Correlation Matrix:")
    print(roads_corr)
    
    plt.figure()
    sns.heatmap(roads_corr, annot=True, cmap="coolwarm", fmt=".2f")
    plt.title("Road Network Data Correlation Matrix")
    plt.show()

# =============================================================================
# 4. Additional Visualizations
# =============================================================================

# ---------------------------
# 4.1 Delivery Data Visualizations
# ---------------------------
# Distribution of delivery times (by hour)
delivery_df = delivery_df.dropna(subset=["delivery_time"])
delivery_df["hour"] = delivery_df["delivery_time"].dt.hour
plt.figure()
sns.countplot(x="hour", data=delivery_df, palette="mako")
plt.title("Hourly Distribution of Deliveries")
plt.xlabel("Hour of Day")
plt.ylabel("Number of Deliveries")
plt.show()

# Delivery locations: Hexbin plot to show density
plt.figure()
plt.hexbin(delivery_df["lng"], delivery_df["lat"], gridsize=50, cmap="viridis", mincnt=1)
plt.colorbar(label="Count")
plt.title("Hexbin Map of Delivery Locations")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()

# ---------------------------
# 4.2 Pickup Data Visualizations
# ---------------------------
pickup_df = pickup_df.dropna(subset=["accept_time"])
pickup_df["hour"] = pickup_df["accept_time"].dt.hour
plt.figure()
sns.countplot(x="hour", data=pickup_df, palette="coolwarm")
plt.title("Hourly Distribution of Pickup Accept Times")
plt.xlabel("Hour of Day")
plt.ylabel("Number of Pickups")
plt.show()

# Plot distribution of time windows (convert time_window_start and time_window_end to minutes)
if {"time_window_start", "time_window_end"}.issubset(pickup_df.columns):
    # Here we assume the times are in a consistent string format; adjust as needed.
    def time_to_min(t_str):
        try:
            t = pd.to_datetime(t_str, format="%m-%d %H:%M:%S")
            return t.hour * 60 + t.minute
        except:
            return np.nan
    pickup_df["tw_start_min"] = pickup_df["time_window_start"].apply(time_to_min)
    pickup_df["tw_end_min"] = pickup_df["time_window_end"].apply(time_to_min)
    # Compute duration of time window
    pickup_df["tw_duration"] = pickup_df["tw_end_min"] - pickup_df["tw_start_min"]
    
    plt.figure()
    sns.histplot(pickup_df["tw_duration"].dropna(), bins=20, kde=True, color="orchid")
    plt.title("Distribution of Pickup Time Window Durations (minutes)")
    plt.xlabel("Time Window Duration (minutes)")
    plt.ylabel("Frequency")
    plt.show()

# ---------------------------
# 4.3 Road Network Visualizations
# ---------------------------
# For roads, besides descriptive stats we can show distribution of road classes
plt.figure()
sns.countplot(data=roads_df, x="fclass", order=roads_df["fclass"].value_counts().index)
plt.title("Distribution of Road Classes")
plt.xlabel("Road Class")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# If available, visualize spatial extent of roads (extracting points from geometry is advanced,
# but we can do a simple approximation by taking the first coordinate from each geometry string)
def extract_first_coord(geom):
    try:
        # Assume geometry is like "LINESTRING (x1 y1, x2 y2, ...)"
        coords = geom.split("(")[1].split(")")[0].split(",")
        first = coords[0].strip().split(" ")
        return float(first[0]), float(first[1])
    except:
        return np.nan, np.nan

roads_df["first_lng"], roads_df["first_lat"] = zip(*roads_df["geometry"].apply(extract_first_coord))
plt.figure()
sns.scatterplot(data=roads_df.dropna(subset=["first_lng", "first_lat"]), x="first_lng", y="first_lat",
                hue="fclass", palette="tab10", legend="full", s=50, alpha=0.8)
plt.title("Approximate Spatial Distribution of Roads")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.legend(title="Road Class", bbox_to_anchor=(1.05, 1))
plt.tight_layout()
plt.show()

# =============================================================================
# 5. Courier-Level Aggregation & Clustering (Delivery Data)
# =============================================================================
# Aggregate delivery data by courier to get centroid and delivery count
if {"courier_id", "lat", "lng", "delivery_time"}.issubset(delivery_df.columns):
    courier_agg = delivery_df.groupby("courier_id").agg({
        "lat": "mean",
        "lng": "mean",
        "delivery_time": "count"
    }).reset_index()
    courier_agg.rename(columns={"delivery_time": "num_deliveries"}, inplace=True)
    print("\nAggregated Courier-Level Data:")
    print(courier_agg.head())

    # Boxplots of delivery counts and lat/lng distributions
    plt.figure()
    sns.boxplot(data=courier_agg, y="num_deliveries", color="lightblue")
    plt.title("Distribution of Deliveries per Courier")
    plt.ylabel("Number of Deliveries")
    plt.show()

    # Scatter plot of courier centroids with delivery counts as size
    plt.figure()
    sns.scatterplot(data=courier_agg, x="lng", y="lat", size="num_deliveries",
                    legend=False, alpha=0.7, color="purple")
    plt.title("Courier Delivery Centroids")
    plt.xlabel("Average Longitude")
    plt.ylabel("Average Latitude")
    plt.show()

    # Clustering on courier-level data
    features = courier_agg[["lat", "lng", "num_deliveries"]]
    scaler = StandardScaler()
    features_scaled = scaler.fit_transform(features)
    
    # Elbow method
    inertias = []
    K_range = range(1, 11)
    for k in K_range:
        km = KMeans(n_clusters=k, random_state=42)
        km.fit(features_scaled)
        inertias.append(km.inertia_)
    plt.figure()
    plt.plot(K_range, inertias, marker="o")
    plt.title("Elbow Method for Optimal K (Couriers)")
    plt.xlabel("Number of Clusters")
    plt.ylabel("Inertia")
    plt.xticks(K_range)
    plt.show()
    
    # Choose k=3 (for example)
    k_optimal = 3
    km = KMeans(n_clusters=k_optimal, random_state=42)
    courier_agg["cluster"] = km.fit_predict(features_scaled)
    plt.figure()
    sns.scatterplot(data=courier_agg, x="lng", y="lat", hue="cluster", palette="Set1",
                    s=100, alpha=0.8)
    plt.title("K-Means Clustering of Couriers")
    plt.xlabel("Average Longitude")
    plt.ylabel("Average Latitude")
    plt.legend(title="Cluster")
    plt.show()

# =============================================================================
# End of Script
# =============================================================================
print("All visualizations generated successfully!")
```

```python
import os
import glob
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# ----------------------------
# 1. Input & Assumptions
# ----------------------------

# Predefined range of pincodes (for our purposes, we assume "region_id" in pickup data equals pincode)
pincode_range = [3, 4, 5]
print("Input pincode range (using region_id as pincode):", pincode_range)

# Assumptions:
# 1. We assume that the 'region_id' column in the pickup CSV files is equivalent to the pincode.
# 2. We assume that higher efficiency is indicated by a higher number of deliveries per unit of average delivery duration.
# 3. We use the aggregated delivery data (from delivery CSV files) to compute performance metrics.
# 4. We use K-Means clustering (with k=3 for simplicity) to separate providers.
print("\nAssumptions:")
print(" 1. 'region_id' in pickup data represents pincode.")
print(" 2. Efficiency score = (number of deliveries) / (average delivery duration).")
print(" 3. K-Means (k=3) will be used to cluster couriers based on efficiency and geography.\n")

# ----------------------------
# 2. Define File Paths & Load Data
# ----------------------------
# Update these paths as needed.
delivery_path = r"C:\Users\mailg\OneDrive\Documents\GitHub\LaDe\LaDe\delivery"
pickup_path   = r"C:\Users\mailg\OneDrive\Documents\GitHub\LaDe\LaDe\pickup"

# Use glob to get all files (we assume CSV files are comma-separated for delivery and pickup)
delivery_files = glob.glob(os.path.join(delivery_path, "delivery_*.csv"))
pickup_files   = glob.glob(os.path.join(pickup_path, "pickup_*.csv"))

print("Found {} delivery file(s):".format(len(delivery_files)))
for f in delivery_files:
    print("  ", f)
print("Found {} pickup file(s):".format(len(pickup_files)))
for f in pickup_files:
    print("  ", f)

# Load all pickup files into one DataFrame
try:
    pickup_df = pd.concat([pd.read_csv(f) for f in pickup_files], ignore_index=True)
    print("\nPickup data loaded successfully. Total records:", len(pickup_df))
except Exception as e:
    print("Error loading pickup files:", e)

# Load all delivery files into one DataFrame
try:
    delivery_df = pd.concat([pd.read_csv(f) for f in delivery_files], ignore_index=True)
    print("Delivery data loaded successfully. Total records:", len(delivery_df))
except Exception as e:
    print("Error loading delivery files:", e)

# ----------------------------
# 3. Preprocess & Aggregate Pickup Data (to get primary region per courier)
# ----------------------------
# We assume pickup_df has columns: 'courier_id', 'region_id'
# Print sample pickup columns for debugging
print("\nPickup Data Columns:", pickup_df.columns.tolist())

# We use the mode of 'region_id' for each courier as their primary region.
try:
    pickup_agg = pickup_df.groupby("courier_id")["region_id"].agg(lambda x: stats.mode(x.dropna())[0][0]).reset_index()
    pickup_agg.rename(columns={"region_id": "primary_region"}, inplace=True)
    print("\nAggregated Pickup Data (Primary Region per Courier):")
    print(pickup_agg.head())
except Exception as e:
    print("Error aggregating pickup data:", e)

# ----------------------------
# 4. Preprocess & Aggregate Delivery Data (for performance metrics)
# ----------------------------
# We assume delivery_df has columns: 'courier_id', 'delivery_time', 'lng', 'lat'
# Convert delivery_time to datetime if needed.
if delivery_df["delivery_time"].dtype == object:
    delivery_df["delivery_time"] = pd.to_datetime(delivery_df["delivery_time"], errors="coerce")

# For each courier, compute:
# - num_deliveries: count of deliveries
# - avg_delivery_duration: for demonstration, we assume a column "delivery_duration" exists; 
#   if not, we simulate it using random numbers.
if "delivery_duration" not in delivery_df.columns:
    # Simulate delivery_duration in minutes (for demo purposes)
    np.random.seed(42)
    delivery_df["delivery_duration"] = np.random.uniform(10, 60, len(delivery_df))
    
delivery_agg = delivery_df.groupby("courier_id").agg({
    "lng": "mean",
    "lat": "mean",
    "delivery_duration": "mean",
    "delivery_time": "count"  # using count as number of deliveries
}).reset_index()
delivery_agg.rename(columns={"delivery_time": "num_deliveries",
                             "delivery_duration": "avg_delivery_duration"}, inplace=True)

# Compute efficiency score (deliveries per minute)
delivery_agg["efficiency_score"] = delivery_agg["num_deliveries"] / delivery_agg["avg_delivery_duration"]

print("\nAggregated Delivery Data (Performance Metrics):")
print(delivery_agg.head())

# ----------------------------
# 5. Merge Aggregated Pickup and Delivery Data
# ----------------------------
try:
    merged_df = pd.merge(delivery_agg, pickup_agg, on="courier_id", how="inner")
    print("\nMerged Courier Data (Delivery performance + Primary Region):")
    print(merged_df.head())
except Exception as e:
    print("Error merging data:", e)

# ----------------------------
# 6. Filter Couriers by Predefined Pincode (region_id) Range
# ----------------------------
# Filter to include only couriers whose primary_region is in the pincode_range.
filtered_df = merged_df[merged_df["primary_region"].isin(pincode_range)].copy()
print("\nNumber of couriers after filtering by pincode (region_id) in {}: {}".format(pincode_range, len(filtered_df)))
print("Filtered Courier Data Sample:")
print(filtered_df.head())

# ----------------------------
# 7. Clustering Analysis on Filtered Courier Data
# ----------------------------
# We'll use features: efficiency_score, num_deliveries, and geographic centroid (lat, lng)
# Prepare feature matrix
features = filtered_df[["efficiency_score", "num_deliveries", "lat", "lng"]].copy()
print("\nFeatures used for clustering:")
print(features.head())

# Scale features
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)

# Use K-Means clustering with a preset number of clusters (e.g., 3)
k_optimal = 3
kmeans = KMeans(n_clusters=k_optimal, random_state=42)
filtered_df["cluster"] = kmeans.fit_predict(features_scaled)
print("\nClustering complete. Cluster assignments (first 10 rows):")
print(filtered_df[["courier_id", "cluster"]].head(10))

# ----------------------------
# 8. Identify Optimal vs Non-Optimal Clusters
# ----------------------------
# We define the optimal cluster as the one with the highest average efficiency_score.
cluster_eff = filtered_df.groupby("cluster")["efficiency_score"].mean().reset_index()
print("\nAverage Efficiency Score per Cluster:")
print(cluster_eff)

optimal_cluster = cluster_eff.sort_values("efficiency_score", ascending=False).iloc[0]["cluster"]
print("\nIdentified Optimal Cluster (highest average efficiency_score):", optimal_cluster)

# Create separate tables for optimal and non-optimal clusters
optimal_df = filtered_df[filtered_df["cluster"] == optimal_cluster].copy()
non_optimal_df = filtered_df[filtered_df["cluster"] != optimal_cluster].copy()

# ----------------------------
# 9. Print Final Results
# ----------------------------
print("\n--- FINAL RESULTS ---\n")
print("Input Pincode Range:", pincode_range)
print("\nTotal Couriers in Predefined Range:", len(filtered_df))
print("\nOptimal 3PL Providers (Cluster {}):".format(optimal_cluster))
print(optimal_df[["courier_id", "primary_region", "num_deliveries", "avg_delivery_duration", "efficiency_score", "lat", "lng", "cluster"]])
print("\nNon-Optimal 3PL Providers (Other Clusters):")
print(non_optimal_df[["courier_id", "primary_region", "num_deliveries", "avg_delivery_duration", "efficiency_score", "lat", "lng", "cluster"]])

# Optionally, save the results to CSV files:
optimal_df.to_csv("optimal_3PL_providers.csv", index=False)
non_optimal_df.to_csv("non_optimal_3PL_providers.csv", index=False)

# ----------------------------
# 10. Additional Debug Plot: Scatter Plot of Couriers by Efficiency
# ----------------------------
plt.figure()
sns.scatterplot(data=filtered_df, x="num_deliveries", y="efficiency_score", hue="cluster", palette="Set1", s=100, alpha=0.8)
plt.title("Courier Efficiency vs. Number of Deliveries")
plt.xlabel("Number of Deliveries")
plt.ylabel("Efficiency Score (deliveries/min)")
plt.legend(title="Cluster")
plt.show()

print("\nProcessing complete. Optimal and non-optimal clusters have been printed and saved to CSV.")
```
